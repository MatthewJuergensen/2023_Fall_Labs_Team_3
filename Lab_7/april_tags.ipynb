```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib import patches

from pupil_apriltags import Detector
```

```python
camera = cv.VideoCapture(0)
ret_val, img = camera.read()
del(camera)

image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(image, 'gray')
```

```python
at_detector = Detector(
   families="tag36h11",
   nthreads=1,
   quad_decimate=1.0,
   quad_sigma=0.0,
   refine_edges=.5,
   decode_sharpening=0.25,
   debug=0
)

tags = at_detector.detect(image, estimate_tag_pose=True, camera_params=[1,1,1,1], tag_size=0.04)

for t in tags:
    print(t.pose_R)
```
