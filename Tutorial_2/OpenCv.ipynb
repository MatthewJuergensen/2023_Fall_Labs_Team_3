```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import patches
```

```python
cv.haveImageReader("stockimg.jpg")
```

```python
#question 1

#read image
img = cv.imread("stockimg.jpg")

#flip image from BGR to RGB so pyplot has proper colors
img = np.flip(img, axis = -1)

#convert to grayscale
grayimg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#show images
plt.subplot(1,2,1)
plt.imshow(img)
#hide graph ticks
plt.xticks([]);
plt.yticks([]);
plt.subplot(1,2,2)
plt.imshow(grayimg, 'gray')
plt.xticks([]);
plt.yticks([]);
```

```python
#question 2

#set parameter values 
lower = 100
upper = 300

#apply canny edge filter
edges = cv.Canny(img, lower, upper)

#show images
plt.subplot(1,2,1)
plt.imshow(img)
plt.xticks([]);
plt.yticks([]);
plt.subplot(1,2,2)
plt.imshow(edges, 'gray')
plt.xticks([]);
plt.yticks([]);
```

```python
#question 3

#load haar cascade
haar_cascade = cv.CascadeClassifier('Haarcascade_frontalface_default.xml')

#detect faces
faces = haar_cascade.detectMultiScale(grayimg, 1.1, 9)

#show img
fig, ax = plt.subplots()
ax.imshow(img)
plt.xticks([]);
plt.yticks([]);

#add rectangle where detected faces are
for (x, y, w, h) in faces:
    rect = patches.Rectangle((x, y), w, h, linewidth = 1, edgecolor = 'r', facecolor = 'none')
    ax.add_patch(rect)
```
